In this projet I will implement The ordinary least squares regession FORMULA .

```python
import matplotlib.pyplot as plt
import numpy as np
print(0)
```

import data from the data.csv file using numpy.genfromtxt() function .

```python
my_data = np.genfromtxt('data.csv', delimiter= ',')
```

Separate the data to x, and to y .

```python
x = my_data[:, 0]
y = my_data[:, 1]
```

Calculate the avreage of x, y by calculating the Sun / the len .

```python
x_bar = int(sum(x) / len(x))
y_bar = int(sum(y) / len(y))
```

The formula of the ordinary last square is :  m = sum((x_i - x_average) * (y_i * y_average)) / sum((x_i - x_average) ^ 2) for i in len(of data) .
Then after calculating m , b = y_average - (m * x_average) .

```python
base = 0
base_2 = 0
for i in range(len(x)):
	base += (x[i] - x_bar) * (y[i] - y_bar)
	base_2 += (x[i] - x_bar)**2
m = base / base_2
b = y_bar - (m * x_bar)
```

her we generate 2 point  to draw our line .  

```python
x_0 = 0
y_0 = m * x_0 + b
x_1 = 1
y_1 = m * x_1 + b
print(x_0)
```

Finally  we present our result in a grafh using matplotlib library .

```python
plt.scatter(x, y, color = 'red')
plt.axline((x_0, y_0), (x_1, y_1))
plt.show()
```
